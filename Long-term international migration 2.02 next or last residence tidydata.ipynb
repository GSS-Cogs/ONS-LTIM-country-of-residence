Long-term international migration 2.02, last or next resident, UK and England and Wales

```python
from gssutils import *
scraper = Scraper('https://www.ons.gov.uk/peoplepopulationandcommunity/populationandmigration/' \
                  'internationalmigration/datasets/longterminternationalmigrationcountryoflastornextresidencetable202')
scraper
```

```python
tab = next(t for t in scraper.distribution().as_databaker() if t.name == 'Table 2.02')
```

```python
observations = tab.excel_ref("B15").expand(RIGHT).filter(contains_string("Estimate")).expand(DOWN).is_not_blank().is_not_whitespace()
CI = observations.shift(RIGHT)
Citizenship1 = tab.excel_ref('B12').expand(RIGHT).is_not_blank()
Citizenship2 = tab.excel_ref('E13').expand(RIGHT).is_not_blank() - tab.excel_ref('AQ14')
Citizenship3 = tab.excel_ref('T14').expand(RIGHT).is_not_blank() - tab.excel_ref('AQ14').expand(RIGHT).is_not_blank()
Geography = tab.excel_ref('A').expand(DOWN).by_index([17,67])
Flow = tab.excel_ref('A').expand(DOWN).by_index([19,35,51,69,85,101])
Year = tab.excel_ref('A12').expand(DOWN) - Geography - Flow - tab.excel_ref('A116').expand(DOWN)  
Tag = tab.excel_ref('B15').expand(RIGHT).is_not_blank()
Revision = tab.excel_ref('AQ14').expand(RIGHT).is_not_blank()
Dimensions = [
            HDim(Year,'Year',DIRECTLY,LEFT),
            HDim(Geography,'Geography',CLOSEST,ABOVE),
            HDim(Citizenship1,'Citizenship1',CLOSEST,LEFT),
            HDim(Citizenship2,'Region',CLOSEST,LEFT),
            HDim(Citizenship3,'Territory',DIRECTLY,ABOVE),
            HDim(Flow,'Flow',CLOSEST,ABOVE),
            HDimConst('Measure Type', 'Count'),
            HDimConst('Unit','People (thousands)'),
            HDim(CI, 'CI',DIRECTLY, RIGHT),
            HDim(Revision,'Revision',DIRECTLY,ABOVE)
            ]
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
def user_perc(x):
    
    if str(x) == ':':
        return '2011 Census Revision'
    else:
        return 'Original Estimate'
    
new_table['Revision'] = new_table.apply(lambda row: user_perc(row['CI']), axis = 1)
```

```python
new_table['Citizenship1'] = new_table['Citizenship1'].str.rstrip('23')
```

```python
def user_perc7(x):
    
    if str(x) == 'European Union2':
        return 'European Union'
    else:
        return x
    
new_table['Region'] = new_table.apply(lambda row: user_perc7(row['Region']), axis = 1)
```
